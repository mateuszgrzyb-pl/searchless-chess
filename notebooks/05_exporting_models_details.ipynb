#### 1. Import libraries.

```python
%cd ..

import keras
from keras.models import load_model

from src.utils.tools import WarmUpCosineDecay
```

#### 2. Procesing CNN & ResNet models.

```python
paths = [
    'models/01_cnn/',
    'models/02_resnet_m/',
    'models/03_resnet_l/',
    'models/04_resnet_xl/',
]
for path in paths:
    model = load_model(path + 'model.keras')  # load model
    model.save_weights(path + 'model.weights.h5')  # save model weights
    config = model.to_json()  # get model hyperparameters
    with open(path + "config.json", "w") as f:
        f.write(config)  # save model hyperparameters
```

#### 3. Processing ViT models.

```python
paths = [
    'models/05_vit_s/',
    'models/06_vit_m/',
]
for path in paths:
    model = load_model(path + 'model.keras', custom_objects={'WarmUpCosineDecay': WarmUpCosineDecay})  # load model
    model.save_weights(path + 'model.weights.h5')  # save model weights
    config = model.to_json()  # get model hyperparameters
    with open(path + "config.json", "w") as f:
        f.write(config)  # save model hyperparameters
```
